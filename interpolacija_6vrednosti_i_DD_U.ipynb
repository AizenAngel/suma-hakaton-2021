```python
def interpolate_misssing_values(df, col):
    for (index, d1) in enumerate(mainMeteoData.values):
        if np.isnan(d1[col]):
            mainMeteoData.iloc[index, col] = (mainMeteoData.iloc[index-1, col] + mainMeteoData.iloc[index+1, col]) / 2
    
    return df
```

```python
def interpolate_missing_strings(df, col):
    for (index, d1) in enumerate(df.values):
        if d1[col] != d1[col]:
            df.iloc[index, col] = df.iloc[index-1, col]
    
    return df
```

```python
def calculate_stats(df, col_name):
    str_data = df[col_name].astype(str)
    m1 = {"digits": [], "letters": [], "combined": []}
    
    
    
    for data in str_data:
        if data.replace(" ", "").replace(",","").replace("-","").isalpha():
            m1["letters"].append(data)
        elif data.replace(" ", "").replace(",","").replace("-","").isdigit():
            m1["digits"].append(data)
        
        else:
            m1["combined"].append(data)
    
    return m1
```
